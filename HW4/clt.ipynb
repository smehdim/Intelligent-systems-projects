```python
import numpy as np
from numpy.random import exponential, binomial
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def central_limit_exponential(s,n):
  means = []
  for i in range(s):
    sample = exponential(0.5,n)
    means.append(np.mean(sample))
  return means
```

```python
means = central_limit_exponential(2000, 100)
```

```python
# plt.hist(means)
sns.histplot(means, kde=True)
plt.xlabel("mean of distributions")
plt.title('distribution of means for exponential distribution')
```

```python
print('mean = {}'.format(np.mean(means)))
print('varince = {}'.format(np.var(means)))
```

```python
def central_limit_binomial(s,n):
  means = []
  for i in range(s):
    sample = binomial(20, 0.8, n)
    means.append(np.mean(sample))
  return means
```

```python
means = central_limit_binomial(2000, 100)
```

```python
sns.histplot(means, kde=True)
plt.xlabel("mean of distributions")
plt.title('distribution of means for binomial distribution')
```

```python
print('mean = {}'.format(np.mean(means)))
print('varince = {}'.format(np.var(means)))
```
